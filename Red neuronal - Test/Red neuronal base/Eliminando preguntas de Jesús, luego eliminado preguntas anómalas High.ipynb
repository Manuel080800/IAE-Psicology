# Se carga el modelo base

```python
import joblib
import dice_ml
import pandas as pd
from sklearn.model_selection import train_test_split

dfbase = pd.read_csv(
    "Dataset_MO_ENG.csv")

## eliminating physical-related questions
dfbase = dfbase.drop(dfbase.columns[102:-1], axis=1)  


## Creando clases 0-> Nivel Bajo, 1->Nivel Medio, 2->Nivel Alto
dicbase = { 1: 0 , 2: 0, 3:1, 4:2, 5:2} 
dfbase['Target'] = dfbase['Target'].map(dicbase)
    
train_colsbase = dfbase.columns[0:-1]
labelbase = dfbase.columns[-1]
Xbase = dfbase[train_colsbase]
ybase = dfbase[labelbase]
target_namesbase=["Low","Medium","High"]

random_state = 13
from imblearn.over_sampling import SMOTE
oversamplebase = SMOTE(random_state=random_state)
Xbase, ybase = oversamplebase.fit_resample(Xbase, ybase)

seedbase = 1
X_trainbase, X_testbase, y_trainbase, y_testbase = train_test_split(Xbase, ybase, test_size=0.33, random_state=seedbase)
mlpbase= joblib.load("modelo_depresion.pkl")
```

```python
dfbase
```

```python
dfbase
```

# Counterfactuals

```python
# Create Data
dbase = dice_ml.Data(dataframe=dfbase, continuous_features=[], outcome_name='Target')


# Create model
mlpbase = joblib.load("modelo_depresion.pkl")
mbase = dice_ml.Model(model=mlpbase, backend="sklearn")


# Create CFs generator using random
expbase = dice_ml.Dice(dbase, mbase, method="random")


# Generate counterfactuals
e1base = expbase.generate_counterfactuals(query_instances=X_testbase[0:2], total_CFs=2, desired_class=2)
e1base.visualize_as_dataframe(show_only_changes=True)
```

# Generando matriz de confusión

```python
from explainerdashboard import ClassifierExplainer, ExplainerDashboard

#removing ´.´ from column names
X_testbase.columns = Xbase.columns.str.replace(r".", "")
explainerbase = ClassifierExplainer(mlpbase, X_testbase, y_testbase)
ExplainerDashboard(explainerbase).run()
```

# Matriz de confusión en el notebook

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

dispbase = ConfusionMatrixDisplay.from_estimator(
        mlpbase,
        X_testbase,
        y_testbase,
        display_labels=['low','med','high'],
        cmap=plt.cm.Blues,
    
    )
    #disp.ax_.set_title(title)

plt.show()
```

# Datos de la matriz de confusión

```python
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_predbase = mlpbase.predict(X_testbase)

print(precision_recall_fscore_support(y_testbase, y_predbase, average='macro'))

print(confusion_matrix(y_testbase,y_predbase))
print('Classification accuracy =',accuracy_score(y_testbase,y_predbase)*100,'%\n')
print(classification_report(y_testbase,y_predbase))
```

# Calculando pendientes de cada pregunta en la clase low y high (1ra Iteración)

```python
# Se obtiene el valor de la pendiente de la clase HIGH:
valoresHYbase = []

limitebase = len(proba_exp_lrbase['ale_values'])

for i in range(0,limitebase):
    valoresHYbase.append([proba_exp_lrbase['ale_values'][i][0][2], proba_exp_lrbase['ale_values'][i][1][2]])
    
pendienteHighbase = []
for i in range(0,limitebase):
    pendienteHighbase.append(valoresHYbase[i][1] - valoresHYbase[i][0])
```

```python
# Se obtiene el valor de la pendiente de la clase LOW:
valoresLYbase = []

limitebase = len(proba_exp_lrbase['ale_values'])

for i in range(0,limitebase):
    valoresLYbase.append([proba_exp_lrbase['ale_values'][i][0][0], proba_exp_lrbase['ale_values'][i][1][0]])
    
pendienteLowbase = []
for i in range(0,limitebase):
    pendienteLowbase.append(valoresLYbase[i][1] - valoresLYbase[i][0])
```

```python
# Se imprime la pendiente y su valor de pregunta:
print("Total de preguntas: " + str(limitebase))
print()
print("Pendientes de la grafica:")
print()
for i in range(0,limitebase):
    print(str(i + 1) + " - Low: " + str(pendienteLowbase[i]))
    print(str(i + 1) + " - High: " + str(pendienteHighbase[i]))
    print()
```

# Análisis de la pendiente y aplicación del umbral

```python
umbral_positivo = 0.01
umbral_negativo = -0.01

limitebase = len(proba_exp_lrbase['ale_values'])
umbralHighbase = []
umbralLowbase = []

for i in range(0,limitebase):
    if pendienteHighbase[i] >= umbral_positivo:
        umbralHighbase.append(1)
        
    if pendienteHighbase[i] <= umbral_negativo:
        umbralHighbase.append(0)
        
    if pendienteHighbase[i] < umbral_positivo and pendienteHighbase[i] > umbral_negativo:
        umbralHighbase.append("NA")
        
    if pendienteLowbase[i] >= umbral_positivo:
        umbralLowbase.append(1)
        
    if pendienteLowbase[i] <= umbral_negativo:
        umbralLowbase.append(0)
        
    if pendienteLowbase[i] < umbral_positivo and pendienteLowbase[i] > umbral_negativo:
        umbralLowbase.append("NA")
```

# re-entrenando el modelo

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np

cv_scores_mean=[]
cv_scores_std=[]

# Entrenar y validar distintas configuraciones
regul_param_range = 10.0 ** -np.arange(-2, 7) # [0.1, 0.01, ..., 0.0000001]
for regul_param in regul_param_range:
    # Aumentamos el parámetro max_iter hasta que converja
    mlp1=MLPClassifier(hidden_layer_sizes=(10,), activation='logistic', solver='adam', alpha=regul_param, 
             learning_rate='constant', learning_rate_init=0.0001, max_iter=100000, random_state=seed1)
    
    # El error es neg_mean_squared_error el MSE en negativo!! Más pequeño peor.
    scores = cross_val_score(mlp1, X1, y1, cv=5, scoring='f1_macro')
    
    cv_scores_mean.append(scores.mean())
    cv_scores_std.append(scores.std())

cv_scores_mean, cv_scores_std
```

```python
import matplotlib.pyplot as plt

# Dibujamos la línea de la exactitud media en las partes de test
plt.plot(np.log10(regul_param_range), cv_scores_mean, color="g", label="Test")

# Dibujamos la banda de la desviación típica
lower_limit = np.array(cv_scores_mean) - np.array(cv_scores_std)
upper_limit = np.array(cv_scores_mean) + np.array(cv_scores_std)
plt.fill_between(np.log10(regul_param_range), lower_limit, upper_limit, color="#DDDDDD")

# Creamos el gráfico
plt.title("Curva de aprendizaje")
plt.xlabel("Alpha 10^{X}"), plt.ylabel("F1"), plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
# Se obtiene el valor de la pendiente de la clase LOW:
valoresLY4 = []

limite4 = len(proba_exp_lr4['ale_values'])

for i in range(0,limite4):
    valoresLY4.append([proba_exp_lr4['ale_values'][i][0][0], proba_exp_lr4['ale_values'][i][1][0]])
    
pendienteLow4 = []
for i in range(0,limite4):
    pendienteLow4.append(valoresLY4[i][1] - valoresLY4[i][0])
```

# Análisis de la pendiente y aplicación del umbral

```python
umbral_positivo = 0.01
umbral_negativo = -0.01

limite4 = len(proba_exp_lr4['ale_values'])
umbralHigh4 = []
umbralLow4 = []

for i in range(0,limite4):
    if pendienteHigh4[i] >= umbral_positivo:
        umbralHigh4.append(1)
        
    if pendienteHigh4[i] <= umbral_negativo:
        umbralHigh4.append(0)
        
    if pendienteHigh4[i] < umbral_positivo and pendienteHigh4[i] > umbral_negativo:
        umbralHigh4.append("NA")
        
    if pendienteLow4[i] >= umbral_positivo:
        umbralLow4.append(1)
        
    if pendienteLow4[i] <= umbral_negativo:
        umbralLow4.append(0)
        
    if pendienteLow4[i] < umbral_positivo and pendienteLow4[i] > umbral_negativo:
        umbralLow4.append("NA")
```
