# Graficación de la explicando ANN (Clasificación) para identificar depresión por fuerza bruta

## Preparación de los datos

Lectura de los datos de acuerdo al dataset de metricas

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("Dataset_QS_METRICS.csv")
```

Ordenación de los datos de acuerdo a la metrica de importancia

```python
df_a = df.sort_values(by = "Acurracy global", ascending = False)
df_p0 = df.sort_values(by = "Precision_0", ascending = False)
df_p1 = df.sort_values(by = "Precision_1", ascending = False)
df_p2 = df.sort_values(by = "Precision_2", ascending = False)
df_r0 = df.sort_values(by = "Recall_0", ascending = False)
df_r1 = df.sort_values(by = "Recall_1", ascending = False)
df_r2 = df.sort_values(by = "Recall_2", ascending = False)
df_f0 = df.sort_values(by = "F1_score_0", ascending = False)
df_f1 = df.sort_values(by = "F1_score_1", ascending = False)
df_f2 = df.sort_values(by = "F1_score_2", ascending = False)
```

## Generación de los gráficos a partir de los datos

Visualización de la gráfica de accurracy global con ordenación

```python
import matplotlib.pyplot as plt

eje_x = df_a.iloc[:, 0].to_numpy().tolist()
eje_y = df_a.iloc[:, 1].to_numpy().tolist()

labels=['Accuracy global']
fig, ax = plt.subplots()

ax.set_xticks(range(len(eje_x)),eje_x)
ax.bar(range(len(eje_y)), eje_y, color = 'tab:purple')

for i,n in enumerate(eje_y):
    ax.text( i, n + 0.5, round(n,4), ha='center', rotation=270, va='bottom', color = 'black')
    
plt.legend(labels)

# ax.grid(axis = 'y', color = 'gray', linestyle = 'dashed')
ax.set_title('Accuracy global', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
fig.set_size_inches(35, 20)
plt.show()
```
